Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Ans:- Definition

Sampling is the process of selecting a representative subset of individuals, items, or observations from a larger population to analyze and draw inferences about the whole group.


Key Terms

Population: The entire group under study (e.g., all students in a university).

Sample: A smaller subset selected from the population.

Sampling Method: The technique used to select the sample.

Sampling Error: The difference between the sample result and the actual population value.

Example

Suppose a researcher wants to study the average income of households in India. Surveying every household is impractical. Instead, the researcher selects a sample of 1,000 households and uses the findings to estimate the national average.

Importance of Sampling

Saves time and resources

Enables quick decision-making

Ensures feasibility in large-scale studies

Improves efficiency in research and data analysis

Random Sampling

Definition

Random sampling, also known as simple random sampling, is a method in which every member of the population has an equal and independent chance of being selected.

Characteristics

Equal probability of selection

Eliminates selection bias

Simple and easy to implement

Requires a complete list of the population
Methods

Lottery method

Random number tables

Computer-generated random selection

Example

A teacher selects 10 students from a class of 50 by drawing names from a box.


Advantages

Unbiased and fair

Easy statistical analysis

Suitable for homogeneous populations

Limitations

May not represent all subgroups

Not ideal for diverse populations

Requires a complete sampling frame

Stratified Sampling

Definition

Stratified sampling involves dividing the population into homogeneous subgroups called strata based on characteristics such as age, gender, income, or education. Samples are then randomly selected from each stratum.

```python
""" Question 5: Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28] """

# Import the statistics module
import statistics

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate mean, median, and mode
mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)

# For datasets with multiple modes, use multimode()
mode_values = statistics.multimode(numbers)

# Display results
print("Numbers:", numbers)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_values)
```

```python
""" Question 6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60] """

import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to NumPy arrays
x = np.array(list_x)
y = np.array(list_y)

# Covariance matrix
cov_matrix = np.cov(x, y)

# Extract sample covariance
sample_covariance = cov_matrix[0, 1]

# Correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Population covariance
population_covariance = np.mean((x - np.mean(x)) * (y - np.mean(y)))

# Display results
print("Sample Covariance:", sample_covariance)
print("Population Covariance:", population_covariance)
print("Correlation Coefficient:", correlation)
```

```python
""" Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35] """

# Import required libraries
import matplotlib.pyplot as plt
import numpy as np

# Given dataset
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Convert data into a NumPy array
data_array = np.array(data)

# Calculate quartiles
Q1 = np.percentile(data_array, 25)
Q2 = np.percentile(data_array, 50)  # Median
Q3 = np.percentile(data_array, 75)

# Calculate Interquartile Range (IQR)
IQR = Q3 - Q1

# Determine lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data_array[(data_array < lower_bound) | (data_array > upper_bound)]

# Print statistical results
print("Q1 (First Quartile):", Q1)
print("Median (Q2):", Q2)
print("Q3 (Third Quartile):", Q3)
print("Interquartile Range (IQR):", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)

# Create the boxplot
plt.figure()
plt.boxplot(data, vert=True, patch_artist=False)
plt.title("Boxplot of the Given Data")
plt.ylabel("Values")
plt.show()
```

```python
""" Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000] """

import numpy as np

# Data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert lists to NumPy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Calculate covariance matrix
cov_matrix = np.cov(x, y)
sample_covariance = cov_matrix[0, 1]

# Calculate population covariance
population_covariance = np.mean((x - np.mean(x)) * (y - np.mean(y)))

# Calculate correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Display results
print("Sample Covariance:", sample_covariance)
print("Population Covariance:", population_covariance)
print("Correlation Coefficient:", correlation)
```
